### Video 1: Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def system(x):
    
    return  2*x**2 - 100*x + 15000
```

$$y = 2x^2-100x+1500$$ 

To calculate the minimum:


$$\frac{\mathrm d}{\mathrm d x} \left( y \right) = 4x-100 = 0$$

$$x = \frac{\mathrm 100}{\mathrm 4}=25$$ 

```python
def drawSys(func):
    xlim = (0,50)
    x = np.linspace(xlim[0],xlim[1],50)
    y = func(x)
    
    ax = plt.subplot()
    ax.plot(x,y)
    plt.show()
```

```python
drawSys(system)
```

```python
def drawSysPoint(func,p):
    xlim = (0,50)
    x = np.linspace(xlim[0],xlim[1],50)
    y = func(x)
    
    ax = plt.subplot()
    ax.plot(x,y)
    ax.scatter(p,func(p),marker='*',color ='C1',s=75)
    ax.axvline(p,linestyle=':',linewidth=0.5,color='gray')
    ax.annotate(text=f'x={p:.2f}',xy=(p,func(1)))
    plt.show()
```

```python
drawSysPoint(system,10)
```

```python
def extract_tangent_line(func,p):
    eps = 0.000001
    x1 = p+eps
    y1 = func(x1)
    
    x2= p-eps
    y2= func(x2)
    
    slope = (y2-y1)/(x2-x1)
    
    intercept = func(p) - slope * p
    
    return slope, intercept, lambda x: intercept+ slope*x
```

```python
extract_tangent_line(system,10)
```

```python
def drawSysPointwTangentLine(func,p):
    xlim = (0,50)
    x = np.linspace(xlim[0],xlim[1],50)
    y = func(x)
    
    ax = plt.subplot()
    ax.plot(x,y)
    ax.scatter(p,func(p),marker='*',color ='C1',s=75)
    ax.axvline(p,linestyle=':',linewidth=0.5,color='gray')
    ax.annotate(text=f'x={p:.2f}',xy=(p,func(1)))
    
    slope,intercept,tangent_line_func = extract_tangent_line(func,p)
    xlim = (p-5,p+5)
    x = np.linspace(xlim[0],xlim[1],50)
    y = tangent_line_func(x)
    ax.plot(x,y,linestyle=':',color='gray')
    
    ax.annotate(text =f'slope:{slope:.2f}',xy=(p+1,func(p)))

    
    plt.show()
```

```python
drawSysPointwTangentLine(system,40)
```

```python
def drawSysPointwTangentLine_proposed_move(func,p):
    xlim = (0,50)
    x = np.linspace(xlim[0],xlim[1],50)
    y = func(x)
    
    ax = plt.subplot()
    ax.plot(x,y)
    ax.scatter(p,func(p),marker='*',color ='C1',s=75)
    ax.axvline(p,linestyle=':',linewidth=0.5,color='gray')
    ax.annotate(text=f'x={p:.2f}',xy=(p,func(1)))
    
    slope,intercept,tangent_line_func = extract_tangent_line(func,p)
    move = -slope*0.15
    xlim = (p-5,p+5)
    x = np.linspace(xlim[0],xlim[1],50)
    y = tangent_line_func(x)
    ax.plot(x,y,linestyle=':',color='gray')
    
    ax.annotate(text =f'slope:{slope:.2f},move:{move:.2f}',xy=(p+1,func(p)))

    
    plt.show()
    return move
```

```python
drawSysPointwTangentLine_proposed_move(system,16.20+5.28)
```

```python
x_candidate = 3
for i in range(20):
    x_candidate += drawSysPointwTangentLine_proposed_move(system,x_candidate)
```
