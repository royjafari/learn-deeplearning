## generate simulated_data.csv

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
positives = np.random.multivariate_normal([1,4], [[1, 0.5], [0.5, 1]], 100000)
positives
```

```python
negatives = np.random.multivariate_normal([4,1], [[1, 0.5], [0.5, 1]], 100000)
negatives
```

```python
x = positives[:,0]
y = positives[:,1]
plt.plot(x, y, 'x')

x = negatives[:,0]
y = negatives[:,1]
plt.plot(x, y, '.')
plt.show()
```

```python
labels = np.append(np.ones(100000),np.zeros(100000))
labels.shape
```

```python
predictors = np.concatenate([positives,negatives])
predictors.shape
```

```python
df = pd.DataFrame(predictors,columns=['X','Y'])
df['label'] = labels.astype(int)
```

```python
(df.sample(frac=1)
 .reset_index(drop=True)
 .to_csv('simulated_data.csv',index=False))
```
